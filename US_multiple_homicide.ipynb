### Combining Categorical Variables

```python
multiple_homicide["Weapon"].value_counts()
```

```python
def combine_categories(df, col, lst, new):
    df[col] = df[col].replace(lst, new)


combine_categories(multiple_homicide, "Weapon", ["Firearm, type not stated", "Other gun", 
                                                 "Rifle", "Shotgun"], "Firearm (Other)")

combine_categories(multiple_homicide, "Weapon", ["Narcotics or drugs, sleeping pills", 
                                                 "Poison - does not include gas"], "Drugs / Poison")

combine_categories(multiple_homicide, "Weapon", ["Asphyxiation - includes death by gas", 
                                                 "Blunt object - hammer, club, etc", 
                                                 "Knife or cutting instrument", 
                                                 "Personal weapons, includes beating", 
                                                 "Strangulation - hanging"], "Close Contact / Blunt Force")

combine_categories(multiple_homicide, "Weapon", ["Handgun - pistol, revolver, etc"], "Handgun")

combine_categories(multiple_homicide, "Weapon", ["Drowning", "Explosives", "Fire", "Other or type unknown", 
                                                 "Pushed or thrown out window"], "Other")
```

```python
# Combine some relationships (mostly family)
# relationship between victim and offender

combine_categories(multiple_homicide, "Relationship", ["Wife", "Girlfriend", "Husband", "Boyfriend", 
                                                       "Common-law wife", "Common-law husband", 
                                                       "Homosexual relationship"], "Current Romantic")

combine_categories(multiple_homicide, "Relationship", ["Ex-wife", "Ex-husband"], "Past Romantic")

combine_categories(multiple_homicide, "Relationship", ["Other family", "In-law", "Stepson", "Stepmother",
                                                       "Stepdaughter", "Stepfather"], "Other Family")

combine_categories(multiple_homicide, "Relationship", ["Sister", "Brother", "Mother", "Father", "Son", 
                                                       "Daughter"], "Immediate Family")

combine_categories(multiple_homicide, "Relationship", ["Employee", "Employer"], "Work")

combine_categories(multiple_homicide, "Relationship", ["Relationship not determined"], "Unknown")

#combine_categories(multiple_homicide, "Relationship", ["Neighbor"], "Other - known to victim")
```

```python
# Group states into regions
northeast = ["ME", "VT", "NY", "PA", "NJ", "NH", "CT", "RI", "MA"]

midwest = ["WI", "ND", "SD", "OH", "MI", "IA", "IL", "MN", "NE", "KS", "MO", "IN"]

west = ["WA", "CA", "OR", "AZ", "NV", "NM", "MT", "CO", "ID", "WY", "AK", "HI", "UT"]

south = ["TX", "LA", "OK", "TN", "AL", "KY", "MS", "GA", "FL", "VA", "WV", "AR", "MD", "DE", 
         "District of Columbia", "NC", "SC"]

# Function to create region columns
multiple_homicide["State"] = multiple_homicide["State"].apply(lambda x: x.strip())

#def set_region(statelist, colname):
    #multiple_homicide[colname] = 0
    #multiple_homicide.loc[multiple_homicide["State"].isin(statelist), colname] = 1
```

```python
# Create and combine region columns
multiple_homicide["Region"] = "0"

def set_region(statelist, region):
    multiple_homicide.loc[multiple_homicide["State"].isin(statelist), "Region"] = region

set_region(northeast, "Northeast")
set_region(south, "South")
set_region(west, "West")
set_region(midwest, "Midwest")
```
